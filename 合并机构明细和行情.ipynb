```python
import tushare as ts
import pandas as pd
df1 = ts.inst_detail()
df1.columns=["股票代码","名称","日期","买入额","卖出额","方法：ts.inst_detail()"]

df2 = ts.get_today_all()
df2.columns = [ "股票代码", "名称", "涨跌幅","现价",
    "开盘价", "最高价", "最低价","昨日收盘价",
    "成交量", "换手率",  "成交量",
    "市盈率", "市净率", "总市值","流通市值"]
df3 = pd.merge(df1,df2,on="股票代码",how="inner")
print(df3)
df3.to_excel("/home/wangd/合并机构明细和行情.xlsx")
```
