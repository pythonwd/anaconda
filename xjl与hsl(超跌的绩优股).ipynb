```python
import pandas as pd
import tushare as ts
import datetime
zs = ts.get_index()
zs = zs.iloc[0,2]#大盘指数
df1 = ts.get_cashflow_data(2018,3)#现金流好的
#df.columns = ["股票代码","名称","经营现金净流量对销售收入比率","资产的经营现金流量回报率"
#,"经营现金净流量与净利润的比率","经营现金净流量对负债比率","现金流量比率"]
df1 = df1.sort_values(by="rateofreturn",ascending=False)
df1 = df1.head(1000)

df2 = ts.get_today_all()
#df2.columns = [ "股票代码", "名称", "涨跌幅","现价",
 #   "开盘价", "最高价", "最低价","昨日收盘价",
 #   "成交量", "换手率",  "成交量",
  #  "市盈率", "市净率", "总市值","流通市值"]
df2 = df2.query("changepercent<%s-2" % zs)#设定当前价格的选股标准,也可以设置成抗跌的股票
df3 = pd.merge(df1,df2,on="code",how="inner")

df4 = ts.get_report_data(2018,3)#考察盈利状况
df4 = df4.query("roe>5.00 and profits_yoy>2.00")#筛选增长性好的

df5 = pd.merge(df3,df4,on="code",how="inner")
df5 = df5.drop_duplicates()
df5.to_excel("/home/wangd/xjl与hsl（超跌的绩优股）.xlsx")

code = df5["code"]
today = datetime.date.today()
oneday = datetime.timedelta(days=20)
beging = str(today-oneday)
for i in code:  #逐个下载历史行情，从技术形态层面分析
    try:
   
        df = ts.get_hist_data('%06s' % i,start='%s' % beging)   #%06d,表示数字是6位数，如不够就自动加0到前面
   
    
        if df.iloc[0,2] < df["close"].mean():#当日收盘价小于历史最高价10%
            
    
            print(i)
            
           
            
        else:
            pass
    except:
        pass



```
