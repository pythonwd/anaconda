```python
import tushare as ts
import pandas as pd
import datetime
import talib
import numpy as np
import matplotlib.pyplot as plt

today = datetime.date.today()
oneday = datetime.timedelta(days=100)
beging = str(today-oneday)
df = ts.get_hist_data('300019',start='%s' %beging)
df = df.sort_index(ascending=False)
close = df['close'].values
macd = talib.MACD(close, fastperiod=12, slowperiod=26, signalperiod=9)
macd=list(tuple(macd))#元组转换成list
 
macd2 = pd.DataFrame(macd)#列表转换成表格
macd2 = macd2.T
print(macd2)
macd2.plot()
```
